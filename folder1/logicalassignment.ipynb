```python
student_name = input("Enter the student name: ")
maths_marks = int(input("Enter Maths marks:"))
science_marks = int(input("Enter Science marks:"))
english_marks = int(input("Enter English marks:"))
```

```python
if maths_marks < 0 or maths_marks > 100 or science_marks < 0 or science_marks > 100 or english_marks < 0 or english_marks > 100:
    print("Invalid marks entered" )
    exit()
```

```python
Total_marks = maths_marks + science_marks + english_marks
average = Total_marks / 3
Average_percentage = (Total_marks/300) * 100
```

```python
if maths_marks < 40 or science_marks <40 or english_marks < 40:
    status = "FAIL"
else:
    status = "PASS"
```

```python
if average >= 75:
    grade = "A"
elif average >= 60 and average < 75:
    grade = "B"
elif average >= 40 and average < 60:
    grade = "C"
```

```python
print(f"Student_Name:{student_name}")
print(f"Total_marks:{Total_marks}")
print(f"Percentage:{Average_percentage:.2f}")
print(f"Status:{status}")
if status == "PASS":
    print(f"Grade:{grade}")
```

Tuples in Python

```python

tup = ()
print(tup)
```

```python
tup = ("python","tuple")
print(tup)
```

```python
li = [1,2,3,4,5]
print(tuple(li))
```

```python
tup = tuple("python")
print(tup)
```

```python
tup = (5,"welcome", 7.6,True)
print(tup)
```

```python
tup1 = (0,1,2,3)
tup2 = ("python","data")
tup3 = (tup1 + tup2)
print(tup3)
```

```python
tup4 = ("code",) * 3
print(tup4)
```

```python
tup = ("python")
n = 3
for i in range(n):
    tup = (tup,)
    print(tup)
```

```python
tup = tuple("python")
print(tup[4])
```

```python
print(tup[1:4])
print(tup[:3])
```
